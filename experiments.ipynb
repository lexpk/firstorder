```python
p = Problem()
p.declare_function('member', 2)
p.declare_function('h', 3)
p.declare_function('intersection', 3)
p.declare_function('true', 0)
p.declare_function('aIb', 0)
p.declare_function('c', 0)
p.declare_function('aIbIc', 0)
```

```python
p.read_axiom('member(h(X0, X1, X2), X0) = true, member(h(X0, X1, X2), X1) = true, member(h(X0, X1, X2), X2) = true -> intersection(X0, X1, X2) = true')
p.read_axiom('-> member(h(aIb, c, aIbIc), aIbIc) = true')
p.read_conjecture('member(h(aIb, c, aIbIc), aIb) = true, member(h(aIb, c, aIbIc), c) = true -> intersection(aIb, c, aIbIc) = true')
```
